```python
import pandas as pd
```

```python
import numpy as np
```

```python
battles = pd.read_csv('battles.csv')
```

```python
battles.head(5)
```

```python
battles.shape  #it gives number of rows and columns in the dataset
```

```python
battles.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
```

```python
battles.head(5)
```

```python
battles.rename(columns={'defender_1':'primary_defender'},inplace=True)
battles.head(5)
```

```python
battles['attacker_king'].value_counts()   # to get a glimpse of number attacks done till now.
```

```python
battles['location'].value_counts() #to get aglimpse of attacks taken place at diffferent locations
```

```python
import seaborn as sns
from matplotlib import pyplot as plt#Seaborn is a powerful Python library for data visualization based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
```

```python
battles['year'].value_counts()
```

```python
battles['battle_number'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(14,6)})        #to get a visualisation of army size of each attacker. first we set the size of plot
sns.barplot(x='attacker_king', y='attacker_size',data=battles)  # we set the x and y axis
plt.show()
```

```python
sns.set(rc={'figure.figsize':(14,6)})
sns.barplot(x='defender_king', y='defender_size',data=battles
            )
plt.show()
```

```python
battles['battle_type'].value_counts()
```

```python
sns.countplot(x='battle_type',data=battles)
plt.show()
```

```python
sns.countplot(x=battles['attacker_king'],hue=battles['battle_type'])
plt.show()
```
